## Creating the dataframes to plot, these will be our target variables

```python
total_population = ["B01003_001E"]
info = total_population +internet_variables 

county_level_data_nineteen = censusdata.download(src="acs1", year=2019, \
                               geo=censusdata.censusgeo([("state", "25"),\
                                                         ("county", "*")]), \
                               key="", var=info)

column_names = ['population', 'w_internet_sub_or_no', 'w_internet_sub', 'dial_up_only', \
                'broadband', 'cellular_data_plan', \
                'only_cellular_data_plan', 'cable_fiber_dsl', 'only_cable_fiber_dsl', \
               'satellite', 'satellite_only', 'only_other', 'w_internet_no_sub', 'no_internet']

county_level_data_nineteen.columns = column_names

new_indices = []
county_names = []
for index in county_level_data_nineteen.index.tolist():
        new_index = index.geo[0][1] + index.geo[1][1]
        new_indices.append(new_index)
        county_name = index.name.split(',')[0]
        county_names.append(county_name)
        
county_level_data_nineteen.index = new_indices
county_level_data_nineteen['county_name'] = county_names
new_df = county_level_data_nineteen

for i in range(1,14):
    new_df[column_names[i]] = new_df.apply(lambda row: row[column_names[i]]/row['population'], axis = 1)
plt.show()

new_df.head()
```

## Examples of how to plot the data 

```python
fig = ff.create_choropleth(fips=new_df.index, 
                           scope=['Massachusetts'],
                           values=new_df.w_internet_sub_or_no, 
                           title='Total with internet', 
                           legend_title='% with internet')
fig.layout.template = None
fig.show()
```

```python
fig = ff.create_choropleth(fips=new_df.index, 
                           scope=['Massachusetts'],
                           values=new_df.w_internet_sub, 
                           title='Total with internet subscriptions', 
                           legend_title='% with internet subscriptions')
fig.layout.template = None
fig.show()
```

```python
fig = ff.create_choropleth(fips=new_df.index, 
                           scope=['Massachusetts'],
                           values=new_df.dial_up_only, 
                           title='Total with dial up only', 
                           legend_title='% with dial up only')
fig.layout.template = None
fig.show()
```
