For each of these counties, we can create tables that illustrate how many people there are by race in each census tract within the county.

```python
from IPython.display import display

counties = [suffolk_county_data, middlesex_county_data, barnstable_county_data, \
            essex_county_data, norfolk_county_data, berkshire_county_data, plymouth_county_data, \
            bristol_county_data, hampden_county_data, franklin_county_data, worcester_county_data, hampshire_county_data]

county_demographics = []

# create the demographic tables for each county
for county in counties:
    county_data = county
    county_data.columns = demographics_column_names
    new_indices = []
    block_groups = []
    tracts = []
    
    for index in county_data.index.tolist():
        new_index = str(index).split(',')[0].replace("Block Group", "")
        tract = str(index).split(',')[1].replace("Census Tract", "")
        new_indices.append(new_index)
        tracts.append(tract)
        block_groups.append("Block Group " + new_index)
    county_data.index = new_indices
    county_data['mean income'] = round(county_data['total income'].div(county_data['total population']))
    county_data['median_income'] = np.where((county_data.median_income < 0), 'NaN',county_data.median_income)
    county_data['census tract'] = tracts
    county_data['block group'] = block_groups
    county_data['% no internet'] = county_data['no internet access'].div(county_data['total population'])
    county_data.drop(columns=['no internet access'])
    county_data.drop(columns=['total income'])
    county_data = county_data.dropna().sort_values(['% no internet'], ascending=False)
    county_demographics.append(county_data)

county_names = ['Suffolk County', 'Middlesex County', 'Barnstable County', 'Essex County', 'Norfolk County', \
                'Berkshire County', 'Plymouth County', \
                'Bristol County', 'Hampden County', 'Franklin County', 'Worcester County', 'Hampshire County']
for i in range(len(county_demographics)):
    print("Number of People by Race in " + county_names[i] + "\n")
    display(county_demographics[i][:10])
    print("\n")
    
```

Or, we can display the demographics as percents and only highlight the top 10% of block groups 
with the lowest % of internet access

```python
for i, county in enumerate(county_demographics):
    new_race_column_names = ['% ' + race for race in race_column_names]
    for index in range(len(new_race_column_names)):
        county[new_race_column_names[index]] = round(county[race_column_names[index]].div(county['total population'])*100,1)
    county = county.drop(columns = demographics_column_names)
    cols = county.columns.tolist()
    cols = cols[1:3] + cols[3:] + [cols[0]]
    county = county[cols]
    print("Percent Of People By Race and Census Tract in " + county_names[i])
    county['% no internet'] = county['% no internet'].div(.01)
    display(county[:10])
    print("\n")
```
